```python


choice = 1










while choice != 0:
    choice = int(input("Виберіть пункт у меню:\n"
                       "1 - Створити ключ\n"
                       "2 - Створити цифровий підпис\n"
                       "3 - Перевірити підпис\n"
                       "0 - Завершити програму\n"))
    if choice == 1:
        public_filename = input("Введіть ім'я файлу для публічного ключа\n")
        private_filename = input("Введіть ім'я файлу для приватного ключа\n")
        create_key(public_filename, private_filename)
    if choice == 2:
        input_choice = int(input("1 - Ввести текст\n2 - Зчитати з файлу\n"))
        if input_choice == 1:
            message = input("Введіть текст повідомлення\n")
        if input_choice == 2:
            file = input("Введіть назву файлу\n")
            with open(file, "r") as file_txt:
                message = file_txt.read()
        elif input_choice != 1 & input_choice != 2:
            print("Вибір некоректний\n")
            continue
        message = str.encode(message)
        key_filename = input("Введіть ім'я файлу з якого взяти приватний ключ\n")
        signature_filename = input("Введіть ім'я файлу для цифрового підпису\n")
        create_signature(message, key_filename, signature_filename)

    if choice == 3:
        input_choice = int(input("1 - Введіть текст\n2 - Введіть назву файлу\n"))
        if input_choice == 1:
            message = input("Введіть текст повідомлення\n")
        if input_choice == 2:
            file = input("Введіть назву файлу\n")
            with open(file, "r") as file_txt:
                message = file_txt.read()
        elif input_choice != 1 & input_choice != 2:
            print("Вибір некоректний\n")
            continue
        message = str.encode(message)
        key_filename = input("Введіть ім'я файлу з якого взяти публічний ключ\n")
        signature_filename = input("Введіть ім'я файлу з якого взяти цифровий підпис\n")
        validate(message, key_filename, signature_filename)
```
