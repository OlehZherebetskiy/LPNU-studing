```python
def lag_func(a, b, t):
    b = b.shift(periods=1)[1:]
    a = a[1:]
    max_corr = a.corr(b)
    lag = 1
    for i in range(1, t):
        b = b.shift(periods=1)[1:]
        a = a[1:]
        t = a.corr(b)
        if t > max_corr:
            lag = i + 1
            max_corr = t
    return lag, max_corr
```

```python
a = df1['Вінницька']
b = df1['Кіровоградська']
lag_func(a, b, 10)
```

```python
 lags = pd.DataFrame()
corrs = pd.DataFrame()

for c1 in df1.columns:
    lag = []
    corr = []
    for c2 in df1.columns:
        res = lag_func(df1[c1], df1[c2], 30)
        lag.append(res[0])
        corr.append(res[1])
    lags[c1] = lag
    corrs[c1] = corr
```

```python
def error(y, y_pred):
    return sum(abs(y_pred[i] - y[i]) / y[i] for i in range(len(y))) / len(y)
```

```python
errors={}
for city in df1.columns:
    lag = lags['Харківська'][city]
    x = df1[['Харківська']][lag:-7]
    y = df1[city][:-7-lag]
    reg = LinearRegression().fit(x, y)
    res = reg.predict(df1[['Харківська']][-7:])
    plt.plot(list(y.values) + list(res), label=city)
    plt.plot(df1[['Харківська']], label='Харківська')
    plt.xticks([])
    plt.title(city)
    plt.legend()
    plt.show()
    
    plt.plot(list(res), label=city)
    plt.plot(list(df1[[city]][-7:].values), label=city+' - real')
    plt.xticks([])
    plt.title(city)
    plt.legend()
    plt.show()
    
    errors[city] = error(df1[city][-7:], res)
    print(errors[city])
print(sorted(errors.items(), key=lambda item: item[1], reverse=True))
print("SUM error: " + str(sum(errors.values())))
print("MAX error: " + str(max(errors.values())))
print("MIN error: " + str(min(errors.values())))
```

```python
# shift + dropna
errors={}
for city in df1.columns:
    if city not in ['Житомирська', 'Хмельницька', 'Харківська']:
        df_zh = df1['Житомирська'].shift(lags[city]['Житомирська'])
        df_hm = df1['Хмельницька'].shift(lags[city]['Хмельницька'])
        df_khr = df1["Харківська"].shift(lags[city]['Харківська'])
        X = pd.concat([df_zh, df_hm, df_khr, df1[city]], axis=1)
        X.dropna(inplace=True)
        y = X[city]
        X.drop(city, axis=1, inplace=True)

        reg = LinearRegression().fit(X[:-7], y[:-7])
        res = reg.predict(X[-7:])
        
        plt.plot(X['Харківська'], label='Харківська')
        plt.plot(X['Житомирська'], label='Житомирська')
        plt.plot(X['Хмельницька'], label='Хмельницька')
        plt.plot(y.index, list(y[:-7].values) + list(res), label=city)
        plt.xticks([])
        plt.title(city)
        plt.legend()
        plt.show()
        
        plt.plot(y.index[-7:], list(res), label=city)
        plt.plot(y.index[-7:], list(y[-7:].values), label=city+" - real")
        plt.xticks([])
        plt.title(city)
        plt.legend()
        plt.show()
        
        errors[city] = error(y[-7:].values, res)
        print(errors[city])
print(sorted(errors.items(), key=lambda item: item[1], reverse=True))
print("SUM error: " + str(sum(errors.values())))
print("MAX error: " + str(max(errors.values())))
print("MIN error: " + str(min(errors.values())))
```

```python
errors={}
data=[]
city = "Чернігівська"
dataCity = ['Житомирська', 'Хмельницька', 'Харківська']
dataCity = df1.columns

for ci in df1.columns:
    if ci != city and ci in dataCity:
        data.append(df1[ci].shift(lags[city][ci]))
data.append(df1[city])
X = pd.concat(data, axis=1)


X.dropna(inplace=True)
y = X[city]
X.drop(city, axis=1, inplace=True)

reg = LinearRegression().fit(X[:-7], y[:-7])
res = reg.predict(X[-7:])

plt.plot(y.index[-7:], list(res), label=city)
plt.plot(y.index[-7:], list(y[-7:].values), label=city+" - real")
plt.xticks([])
plt.title(city)
plt.legend()
plt.show() 

print("error: " + str(error(y[-7:].values, res)))

```
