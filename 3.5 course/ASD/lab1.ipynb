```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel('phones.xlsx')
```

```python
df
```

```python
def productByClientId(data, id):
    return data[data['User ID']== id]
```

```python
productByClientId(df, 200)
```

```python
sortedDf = df.sort_values(by='Sales Price', ascending=False)
sortedDf.head(5)
```

```python
df['User ID'].value_counts().nlargest(5).to_frame()
```

```python
a = df['User ID'].value_counts().nlargest(5).to_frame().reset_index().rename(columns={'User ID': 'count'}).rename(columns={'index': 'User ID'})
topDf =  pd.merge(df,a,on = "User ID")
topDf
```

```python
df[['User ID', 'Sales Price']].groupby(['User ID']).sum().sort_values(by=['Sales Price'], ascending=False).head(15)
```

```python
sns.set_theme(style="whitegrid")
```

```python
sns.barplot(x=topDf['Date'].dt.year, y = 'Sales Price', data = topDf, estimator=sum, hue = 'User ID', ci = None)
```

```python
sns.lineplot(x=df['Date'].dt.year, y = 'Sales Price', data = df, estimator=sum, ci = None)
```
